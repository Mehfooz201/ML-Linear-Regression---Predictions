```python
home_price = pd.read_excel('data/home_price.xlsx')
```

```python
home_price
```

```python
plt.xlabel('Area')
plt.ylabel('Price')
plt.scatter(home_price.area, home_price.price, color='red', marker='+')
```

```python
new_data = home_price.drop('price', axis='columns')
new_data
```

```python
model = linear_model.LinearRegression()
model.fit(new_data, home_price.price) #we call fit to train the model
```

```python
model.predict([[3300]])
```

```python
model.predict([[5000]])
```

# Exercise
Predict canada's per capita income in year 2020. There is an exercise folder here on github at same level as this notebook, download that and you will find canada_per_capita_income.csv file. Using this build a regression model and predict the per capita income fo canadian citizens in year 2020

# Answer

```python
canada_income = pd.read_csv('https://raw.githubusercontent.com/codebasics/py/master/ML/1_linear_reg/Exercise/canada_per_capita_income.csv')
canada_income

#array([[41288.69409442]])
```

```python
#41288.69409442
canada_income.head()
```

```python
canada_income = canada_income.rename(columns={'per capita income (US$)': 'per_capita_income'})
canada_income
```

```python
model = linear_model.LinearRegression()
model.fit(canada_income[['year']], canada_income[['per_capita_income']])
```

```python
model.predict([[2020]])
```
